```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import os

# Data Augmentation settings
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load the input image
img = load_img('copy_1.jpg')  # Make sure this image is in your working directory
x = img_to_array(img)         # Convert image to array
x = x.reshape((1,) + x.shape) # Reshape for ImageDataGenerator

# Output directory
output_dir = 'augmented_output'
os.makedirs(output_dir, exist_ok=True)

# Generate and save 10 images
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir=output_dir, save_prefix='aug', save_format='jpg'):
    i += 1
    if i >= 10:
        break
```

```python
import matplotlib.pyplot as plt
import glob
import matplotlib.image as mpimg

# Show augmented images
image_paths = glob.glob('augmented_output/*.jpg')
plt.figure(figsize=(12, 6))
for i, path in enumerate(image_paths[:5]):
    img = mpimg.imread(path)
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```
